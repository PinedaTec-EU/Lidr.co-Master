````python
input_tokens = response.usage.input_tokens
output_tokens = response.usage.output_tokens

# precios aproximados (ejemplo gpt-4o-mini, ajusta si quieres)
price_input_per_1k = 0.00015
price_output_per_1k = 0.0006

cost = (input_tokens / 1000 * price_input_per_1k) + \
       (output_tokens / 1000 * price_output_per_1k)

print(f"Estimated cost: ${cost:.6f}")
````

````python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        {
            "role": "system",
            "content": "You must respond ONLY in valid JSON. No extra text."
        },
        {
            "role": "user",
            "content": "Give me a startup idea about AI and healthcare with name, description and difficulty (easy, medium, hard)."
        }
    ]
)

print(response.output_text)
````

````python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        {
            "role": "system",
            "content": "You must respond ONLY in valid JSON. No extra text."
        },
        {
            "role": "user",
            "content": "Give me a startup idea about AI and healthcare with name, description and difficulty (easy, medium, hard). Be creative and explain your reasoning."
        }
    ]
)

print(response.output_text)
````

````python
clean_text = response.output_text.strip()

if clean_text.startswith("```json"):
    clean_text = clean_text.removeprefix("```json").strip()

if clean_text.endswith("```"):
    clean_text = clean_text.removesuffix("```").strip()

print(clean_text)
````

````python
import json

try:
    data = json.loads(clean_text)
    print("JSON válido")
    print(data)
except Exception as e:
    print("JSON inválido:", e)
````

````python
print("Name:", data["startup_idea"]["name"])
print("Difficulty:", data["startup_idea"]["difficulty"])
print("Description:", data["startup_idea"]["description"])
````

````python
response = client.responses.create(
    model="gpt-4o-mini",
    text={
        "format": {
            "type": "json_schema",
            "name": "startup_idea",
            "schema": {
                "type": "object",
                "properties": {
                    "name": {"type": "string"},
                    "description": {"type": "string"},
                    "difficulty": {"type": "string"}
                },
                "required": ["name", "description", "difficulty"],
                "additionalProperties": False
            }
        }
    },
    input="Give me a startup idea about AI and healthcare."
)

print(response.output_text)
````
